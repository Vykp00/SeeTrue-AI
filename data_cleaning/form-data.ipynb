# The original csv with data is divided by semicolumns, what is not correctly read by pandas library for data processing. We have deleted commas manually and replaced semicolonns on commas, since in pandas comma is the delimiter for columns in text data.

```python
def replace_semicolons(file_path, output_path=None):
    # Open the file and read its content
    with open(file_path, 'r') as file:
        content = file.read()

    # Replace all semicolons with commas
    modified_content = content.replace(';', ',')

    # Define the output file (same as input if not provided)
    if output_path is None:
        output_path = file_path

    # Write the modified content back to the file
    with open(output_path, 'w') as file:
        file.write(modified_content)

    print(f'Successfully replaced semicolons with commas in: {output_path}')
```

```python
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        input_file = os.path.join(dirname, filename)
        replace_semicolons(input_file, filename)
```
