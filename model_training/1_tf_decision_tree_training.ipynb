```python
import ydf  # Yggdrasil Decision Forests
import pandas as pd 
print("Yggdrasil Decision Forests: ",ydf.__version__)
```

```python
import os
```

```python
# Load the dataset
from data_processing.data_path import DATA_SPLIT_PATH

train_ds = pd.read_csv(os.path.join(DATA_SPLIT_PATH,'train.csv'))
test_ds = pd.read_csv(os.path.join(DATA_SPLIT_PATH,'test.csv'))
```

```python
# Print first few cell
train_ds.head(10)
```

## **<font style = "color:rgb(58, 45, 247)"> Model Training </font>**

```python
# Our labeler is 'result'
# Note: ydf.Task.CLASSIFICATION is the default value of "task"
model = ydf.RandomForestLearner(label="result",
                                task=ydf.Task.CLASSIFICATION).train(train_ds)
```

## **<font style = "color:rgb(58, 45, 247)"> Making Prediction </font>**
Classification models predict the probability of the label classes. Binary classification models output the probabiltity of the first class according to model.label_classes().

```python
"""
Classification models are evaluated using accuracy, confusion matrices, ROC-AUC and PR-AUC. You can plot a rich evaluation with ROC and PR plots.
"""
evaluation = model.evaluate(test_ds)

evaluation
```

```python
print(evaluation.accuracy)
```

```python
model.describe()
```

```python
model.analyze(test_ds, sampling=0.1)
```
